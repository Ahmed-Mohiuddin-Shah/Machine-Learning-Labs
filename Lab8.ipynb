### Strengths and Weaknesses in Different Scenarios

#### When to Use Naïve Bayes 
- Large Datasets: Naive Bayes handles larger datasets with ease due to its low 
computational cost. 
- Text Classification: It performs well in text-based applications like spam detection or 
sentiment analysis, where the independence assumption often holds. 
- Feature Independence: If features are mostly independent, Naive Bayes is a great 
choice. 
#### When to Use KNN 
- Small Datasets: KNN is effective for small datasets where instance-based learning is 
manageable. 
- No Assumptions: If the data’s underlying distribution is unknown or complex, KNN’s 
flexibility can be an advantage. 
- Highly Localized Patterns: When classification needs to adapt to very specific 
patterns, KNN is beneficial due to its locally driven decision-making.

## Training Naive Bayes

### We train Naive Bayes with guassian and bernoulli distribution

#### Making the Function

```python
def train_naive_bayes(train_set_x, y_train_set, X_test, y_test, noise):

    # Train Naïve Bayes classifiers with different distributions

    nb_gaussian = GaussianNB()

    nb_bernoulli = BernoulliNB()


    nb_gaussian.fit(train_set_x, y_train_set)

    nb_bernoulli.fit(train_set_x, y_train_set)


    # Make predictions

    y_pred_gaussian = nb_gaussian.predict(X_test)

    y_pred_bernoulli = nb_bernoulli.predict(X_test)


    # Calculate accuracy

    accuracy_gaussian = accuracy_score(y_test, y_pred_gaussian)

    accuracy_bernoulli = accuracy_score(y_test, y_pred_bernoulli)


    # Confusion matrices

    conf_matrix_gaussian = confusion_matrix(y_test, y_pred_gaussian)

    conf_matrix_bernoulli = confusion_matrix(y_test, y_pred_bernoulli)

    print(f"Classsification Report For Noise {noise}: ")
    print("Gaussian Naive Bayes: ")
    print(classification_report(y_test, y_pred_gaussian))
    print("Bernoulli Naive Bayes: ")
    print(classification_report(y_test, y_pred_bernoulli))


    return accuracy_gaussian, accuracy_bernoulli, conf_matrix_gaussian, conf_matrix_bernoulli, nb_gaussian, nb_bernoulli
```

## Training kNN

### Creating Function to train kNN classifier

```python
def train_knn(train_set_x, y_train_set, X_test, y_test, noise):
    knn_models = []

    # Train KNN models with different values of K
    for k in range(1, 20):
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(train_set_x, y_train_set)
        knn_models.append(knn)

    # Make predictions
    y_preds = [knn.predict(X_test) for knn in knn_models]

    # Calculate accuracies
    accuracies = [accuracy_score(y_test, y_pred) for y_pred in y_preds]

    # Find the best K
    best_k = np.argmax(accuracies) + 1

    best_model = KNeighborsClassifier(n_neighbors=best_k)
    best_model.fit(train_set_x, y_train_set)

    y_preds_best = best_model.predict(X_test)

    # Confusion matrix for the best K
    conf_matrix = confusion_matrix(y_test, y_preds[best_k - 1])

    print(f"Classification Report For Best K for Noise {noise}")
    print(classification_report(y_test, y_preds_best))

    return accuracies, conf_matrix, best_k, best_model
```

### Creating Function to Plot Decision Boundary

```python
# Creating Function to Plot Decision Boundary
def plot_decision_boundary(X, y, model, title):
    h = .02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', s=50)
    plt.xlabel("feature 1")
    plt.ylabel("feature 2")
    plt.title(title)
    # plt.show()
```

```python
def voting_model_maker(results, kNN_results, voting_type, noise, X_train, y_train, X_test, y_test, X, y):
    # Hard voting
    voting_model = VotingClassifier(
        estimators=[("nb", results[4]), ("knn", kNN_results[3])], voting=voting_type
    )
    voting_model.fit(X_train, y_train)

    voting_predictions = voting_model.predict(X_test)
    voting_accuracy = accuracy_score(y_test, voting_predictions)
    print("Noise = 0.1")
    print("Gaussian Accuracy:", results[0])
    print("Accuracy: ", kNN_results[0][(kNN_results[2] - 1)])
    print(f"Hard Voting Accuracy: {voting_accuracy}")

    print(classification_report(y_test, voting_predictions))

    plot_decision_boundary(
        X, y, voting_model, f"{voting_type} Voting Model for {noise} Noise"
    )
    plt.show()
```
